```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('gas_prices.csv')
```

```python
df
```

```python
df.describe()
```

```python
import seaborn as sns
```

```python
# Extracting years and gas prices for a specific country (e.g., USA)
years = df['Year']
gas_prices_usa = df['USA']

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(years, gas_prices_usa, marker='o', linestyle='-')

# Adding labels and title
plt.title('Average Gas Prices in USA')
plt.xlabel('Year')
plt.ylabel('Average Gas Price ($/gallon)')

# Adding gridlines
plt.grid(True)

# Displaying the graph
plt.show()

#
```

```python
# Selecting only the columns with gas prices for different countries
countries_gas_prices = df.drop(columns=['Year'])

# Calculating the correlation matrix
correlation_matrix = countries_gas_prices.corr()

# Creating the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Gas Prices in Different Countries')
plt.xlabel('Countries')
plt.ylabel('Countries')
plt.show()
```

```python
country = 'USA'
gas_prices_country = df[country]

# Creating the distribution plot
plt.figure(figsize=(4, 4))
sns.histplot(gas_prices_country, kde=True, color='skyblue')
plt.title(f'Distribution of Gas Prices in {country}')
plt.xlabel('Gas Price ($/gallon)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
def distplot(data, x, ax):
    sns.kdeplot(data[x], ax=ax, color='skyblue')

# Assuming df is your pandas DataFrame containing the gas price data
# Selecting a specific country (e.g., USA) for visualization
country = 'USA'

# Creating the plot
fig, ax = plt.subplots(figsize=(4, 4))
distplot(df, country, ax)
ax.set_title(f'Distribution of Gas Prices in {country}')
ax.set_xlabel('Gas Price ($/gallon)')
ax.set_ylabel('Density')
ax.grid(True)
plt.show()
```

```python
country = 'USA'

# Creating the catplot
sns.catplot(x='Year', y=country, kind='swarm', data=df, height=6, aspect=2)
plt.title(f'Distribution of Gas Prices in {country} by Year')
plt.xlabel('Year')
plt.ylabel('Gas Price ($/gallon)')
plt.grid(True)
plt.show()
```

```python

# Selecting a specific country (e.g., USA) for visualization
country = 'USA'

# Creating the lmplot
sns.lmplot(x='Year', y=country, data=df, height=6, aspect=2)
plt.title(f'Relationship between Year and Gas Price in {country}')
plt.xlabel('Year')
plt.ylabel('Gas Price ($/gallon)')
plt.grid(True)
plt.show()
```

```python
gas_prices_countries = df.drop(columns=['Year'])

# Creating the pair plot
sns.pairplot(gas_prices_countries)
plt.suptitle('Pair Plot of Gas Prices Across Different Countries', y=1.02)
plt.show()
```

```python
gas_prices_countries = df.drop(columns=['Year'])

# Creating the PairGrid
g = sns.PairGrid(gas_prices_countries)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot, kde=True)

# Adding titles and adjusting layout
plt.suptitle('Pair Grid of Gas Prices Across Different Countries', y=1.02)
plt.tight_layout()
plt.show()
```

```python

# Assuming df is your pandas DataFrame containing the gas price data
# Selecting all columns except 'Year' for the violin plot
gas_prices_countries = df.drop(columns=['Year'])

# Creating the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=gas_prices_countries, inner="quartile")
plt.title('Violin Plot of Gas Prices Across Different Countries')
plt.xlabel('Country')
plt.ylabel('Gas Price ($/gallon)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

import pandas as pd

# Selecting countries for visualization
countries = ['USA', 'Canada', 'Germany', 'Japan']

# Plotting gas prices for each country
plt.figure(figsize=(10, 6))
for country in countries:
    plt.plot(df.index, df[country], label=country)

plt.title('Gas Prices Over Time')
plt.xlabel('Year')
plt.ylabel('Gas Price ($/gallon)')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Plotting the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=gas_prices_countries)
plt.title('Box Plot of Gas Prices Across Different Countries')
plt.ylabel('Gas Price ($/gallon)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
